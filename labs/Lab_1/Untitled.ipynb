# Max element and span in a list

```python
%run max_in_list.py
```

```python
from random import seed, randrange

for_seed = input('Input a seed for the random number generator:')
nb_of_elmt = input('How many elements do you want to generate?')
seed(int(for_seed))
elmt = [randrange(0,100) for _ in range(int(nb_of_elmt))]
print(f'The list is: {elmt}')
min_elmt = float('inf')
max_elmt = -float('inf')
for e in elmt:
    if e < min_elmt:
        min_elmt = e
    if e > max_elmt:
        max_elmt = e
span = max_elmt - min_elmt
print(f'The maximum difference between largest and smallest values in this list is: {span}')
print(f'Confirming with builtin operations: {max(elmt)-min(elmt)}')
```

# Classifying elements in a list

```python
%run modulo_4.py
```

```python
from random import seed, randrange

for_seed = int(input('Input a seed for the random number generator:'))
nb_of_elmt = int(input('How many elements do you want to generate?'))
seed(for_seed)
elmt = [ randrange(0,20) for _ in range(nb_of_elmt)]
print(f'The list is: {elmt}')
zero_four,five_nine,ten_forteen, fifteen_nineteen = 0,0,0,0
for e in elmt:
    if e < 5:
        zero_four += 1
    elif e < 10:
        five_nine += 1
    elif e < 15:
        ten_forteen += 1
    else:
        fifteen_nineteen += 1
if zero_four == 0:
    print(f'There is no element between 0 and 4.')
elif zero_four == 1:
    print(f'There is 1 element between 0 and 4.')
else:
    print(f'There are {zero_four} elements between 0 and 4.')

if five_nine == 0:
    print(f'There is no element between 5 and 9.')
elif five_nine == 1:
    print(f'There is 1 element between 5 and 9.')
else:
    print(f'There are {five_nine} elements between 5 and 9.')

if ten_forteen == 0:
    print(f'There is no element between 10 and 14.')
elif ten_forteen == 1:
    print(f'There is 1 element between 10 and 14.')
else:
    print(f'There are {ten_forteen} elements between 10 and 14.')     

if fifteen_nineteen == 0:
    print(f'There is no element between 15 and 19.')
elif fifteen_nineteen == 1:
    print(f'There is 1 element between 15 and 19.')
else:
    print(f'There are {fifteen_nineteen} elements between 15 and 19.')     
```

# Statistics on numbers in a list

```python
help()
```

```python
from statistics import mean, median, pstdev
from random import seed, randrange

for_seed = int(input('Input a seed for the random number generator:'))
nb_of_elmt = int(input('How many elements do you want to generate?'))
seed(for_seed)
elmt = [randrange(-50,51) for _ in range(nb_of_elmt)]

print(f'The list is:{elmt}')

mmean = round(sum(elmt)/nb_of_elmt,2)
sorted_elmt = sorted(elmt)
if nb_of_elmt%2 == 0:
    mmedian = round((sorted_elmt[nb_of_elmt/2 - 1]+sorted_elmt[nb_of_elmt/2])/2,2)
else:
    mmedian = round(sorted_elmt[nb_of_elmt//2],2)
mean_diff_sum = 0
for e in elmt:
    mean_diff_sum += (e-mmean)**2
st_dev = round((mean_diff_sum/nb_of_elmt)**0.5,2)
print(f'The mean is: {mmean}.')
print(f'The median is: {"{:.2f}".format(mmedian)}.')
print(f'The standard deviation is: {st_dev}.')
print('Confirming with functions from the statistics module:')
print(f'The mean is: {round(mean(elmt),2)}.')
print(f'The median is: {"{:.2f}".format(median(elmt))}')
print(f'The standard deviation is: {round(pstdev(elmt),2)}.')
```
